```python
new_df=df.copy()
new_df
```

```python
# new_df['TotalCharges'].astype(float)
new_df['TotalCharges'].unique()
```

```python
new_df['TotalCharges']=pd.to_numeric(new_df['TotalCharges'], errors='coerce')
```

```python
#here 11 null values appear after converting it into int type
null=new_df['TotalCharges'].isnull().sum()
tot_len=len(new_df)
```

```python
# only 0.15 percentage of the data as null values it would be better
# to delete the data
null/tot_len
```

```python
#removing the missing values
df.dropna(how='any',inplace=True)
```

```python
new_df['tenure'].describe()
```

```python
#here we can divide the tenure into groups
labels=['{0} - {1}'.format(i,i+11)for i in range(1,72,12)]
labels
```

```python
new_df['tenure_group']=pd.cut(new_df.tenure,range(1,80,12),right=False,labels=labels)
```

```python
new_df['tenure_group'].value_counts()
```

```python
#removing cols (customerid and tenure)
new_df.drop(columns=['customerID','tenure'],axis=1,inplace=True)
```

```python
new_df
```

 # univariate analysis

```python
for i,predictor in enumerate(new_df.drop(columns=['MonthlyCharges','TotalCharges','Churn'])):
    plt.figure(i)
    sns.countplot(data=new_df,x=predictor,hue='Churn')
```

```python
new_df['SeniorCitizen'].value_counts()/len(new_df)*100
```

```python
for column in new_df.columns.drop(['MonthlyCharges','TotalCharges','Churn']):
    print(new_df[column].value_counts()/len(new_df)*100)
```

```python
new_df['Churn']=np.where(new_df.Churn=='Yes',1,0)
```

```python
#converting all the categorical variables into dummy variable
new_df_dummies=pd.get_dummies(new_df)
new_df_dummies.head()
```

```python
new_df_dummies['MonthlyCharges'].corr(new_df_dummies['TotalCharges'])
```

```python
mth=sns.kdeplot(new_df_dummies.MonthlyCharges[(new_df_dummies['Churn']==0)],
               color='Red',shade=True)
mth=sns.kdeplot(new_df_dummies.MonthlyCharges[(new_df_dummies['Churn']==1)],
               color='Blue',shade=True)
```

 Churn is high when monthly charges are high and vice versa

```python
plt.figure(figsize=(30,15))
new_df_dummies.corr()['Churn'].sort_values(ascending=False).plot(kind='bar')
```
